```python
import numpy as np
vector = np.array([1, 2, 3]) # crea un vector de numpy constante
print(vector)
print("version  ", np.__version__)
print("tipo de dato  ", type(vector))# tipo de dato
print("dimension  ", vector.ndim) # dimension del vector
print("forma  ", vector.shape) # forma del vector
#crear vector de ceros
vector_ceros = np.zeros(5)
print("vector de ceros  ", vector_ceros)
#crear vector para llenar con unos
vector_unos = np.ones(5)
print("vector de unos  ", vector_unos)
#crear vector con un rango de numeros
vector_rango = np.arange(10) # crea un vector con un rango de numeros
print("vector con rango  ", vector_rango)

```

**Operaciones con Vectore**

```python
import numpy as np
vector = np.arange(20) # rango del 1 al 20
vector1=np.random.rand(1,20) # que se llene de valores aleatorios
print("rango",vector)
print("se llena aleatoriamente numeros reales",vector1)

vector2=np.random.randint(1,100,10)# llena numero aleatorios del 1 al 10 pero de un rango del 1 al 100 como si fuera un baloto
print("aleatoria enteros",vector2)
print("tamaño", vector2.shape)
# suma de vectores
resultadovector = vector + vector1
print("suma de vectores",resultadovector)
# multiplicacion de vectores
resultadovector1 = vector * vector1
print("multiplicacion de vectores",resultadovector1)
# division de vectores
resultadovector2 = vector / vector1
print("division de vectores",resultadovector2)
```
